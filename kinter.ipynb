```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("Student Mental health.csv")

# Create a line graph
plt.figure(figsize=(10, 6))
plt.plot(data["Timestamp"][0:30], data["What is your CGPA?"][0:30])
plt.title("CGPA Over Time")
plt.xlabel("Timestamp")
plt.ylabel("CGPA")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Create a bar graph
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x="Your current year of Study")
plt.title("Distribution of Current Year of Study")
plt.xlabel("Current Year of Study")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="Age", y="Do you have Depression?")
plt.title("Scatter Plot of Age vs Depression")
plt.xlabel("Age")
plt.ylabel("Depression")
plt.tight_layout()
plt.show()

# Calculate the correlation between Age and Do you have Anxiety
#age_anxiety_corr = data["Age"].corr(data["Do you have Anxiety?"])
#age_anxiety_corr = data["Age"].astype(int).corr(data["Do you have Anxiety?"].map({"Yes": 1, "No": 0}))

# Create a correlation matrix just for Age and Do you have Anxiety
#age_anxiety_corr_matrix = pd.DataFrame(data={"Age": [1, age_anxiety_corr], "Do you have Anxiety?": [age_anxiety_corr, 1]})

# Create a correlation heatmap for Age and Do you have Anxiety
#plt.figure(figsize=(6, 4))
#sns.heatmap(age_anxiety_corr_matrix, annot=True, cmap="coolwarm", fmt="d")
#plt.title("Correlation Heatmap: Age vs. Do you have Anxiety?")
#plt.tight_layout()
#plt.show()









```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Student Mental health.csv")

df['Anxiety'] = df['Do you have Anxiety?'].map({'Yes': 1, 'No': 0})

subset_df = df[['Age', 'Anxiety']]

correlation_matrix = subset_df.corr()

plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True, annot_kws={"size": 12})
plt.title('Correlation Heatmap: Age vs. Anxiety')
plt.show()
```
